#### Q1. Explain GET and POST methods.

#### What is a Get Request?

1. Say, i want to browse YouTube and look for videos belonging PWSKILLS python.
  so i will search in youtube saying --> “PWSKILLS python”

https://www.youtube.com/results?search_query=PWSKILLS+python+

2. The end part of the URL that consists of results?search_query=PWSKILLS+python+ is a parameter that is sent to the YouTube server to get videos related to the corresponding search query i.e. PWSKILLS python. 

3. This is an example of GET request. We are requesting data and according to our request, backend server responses.

4. Once making a Get request and server responses with a status code of 200 (which means the request is accepted by the server and the server responded with a positive response)

#### What is a Post Request?

1. a Post request is used to send data to the server to create or update resources. For that, we make a Post request to the server, and that data will be added to the database in the ideal case.

2. To relate this with a real-life example, we use social media platforms and there we see so many posts on our feed (through get request) and billions of posts are added on a daily basis (through post request).

3. In Post request, data is not visible on the URL like GET request. Instead, we pass our data inside the body of POST request



|Diff b/w | Get |Post |
|---|---|---|
|1.| Data is visible on the GET request URL.	| Data is Not visible in the request so you can pass sensitive data like passwords etc.|
|2.| Data length should be maintained to avoid exceeding URL length limit.	|No limit on data length is there in POST request.|
|3.|It supports only string data types	|It supports different data types like strings, boolean, integer, etc|
|4.|Get request can be bookmarked|	Post request can’t be bookmarked|
|5.|Get is simple to use because of its nature of appending data to URL only.	|Post requires header information, body, etc which makes it hard to use as compared with Get request.|
|6.|Get requestsrequest can be cached.	| POST requestsrequest can’t be cached.|

#### Q2. Why is request used in Flask?

1. The request object in Flask is used to access the data that is sent from the client to the server. 

2. This data can include the following:

        The HTTP method (GET, POST, PUT, DELETE)
        The URL
        The headers
        The body of the request

3. The request object is made available to view functions in Flask. This allows view functions to access the data that is sent from the client and use it to process the request.

```python
Example :

## importing required modules
from flask import Flask, request, render_template, jsonify

app = Flask(__name__)

@app.route('/',methods=['GET','POST'])
def home_page():
    data = request.args.get('x')
    return "this is a data input from the my URL : {}".format(data)


if __name__=="__main__":
    app.run(host="0.0.0.0") 
```

#### Q3. Why is redirect() used in Flask?


1. Flask class has a redirect() function. 

2. When called, it returns a response object and redirects the user to another target location with specified status code.

Prototype of redirect() function is as below −  Flask.redirect(location, statuscode, response)

3. In the above function −

    1. location parameter is the URL where response should be redirected.

    2. statuscode sent to browser’s header, defaults to 302.

    3. response parameter is used to instantiate response.


4. The following status codes are standardized −

    1. HTTP_300_MULTIPLE_CHOICES
    2. HTTP_301_MOVED_PERMANENTLY
    3. HTTP_302_FOUND
    4. HTTP_303_SEE_OTHER
    5. HTTP_304_NOT_MODIFIED
    6. HTTP_305_USE_PROXY
    7. HTTP_306_RESERVED
    8. HTTP_307_TEMPORARY_REDIRECT
    
5. The default status code is 302, which is for ‘found’.

#### Q4. What are templates in Flask? Why is the render_template() function used?

1. Templates are files that contain static data as well as placeholders for dynamic data. 
2. A template is rendered with specific data to produce a final document. 
3. Flask uses the Jinja template library to render templates

**render_template()**
1. render_template is a Flask function from the flask.templating package. render_template is used to generate output from a template file based on the Jinja2 engine that is found in the application's templates folder.

2. Note that render_template is typically imported directly from the flask package instead of from flask.templating


```python
## The below the sample snippet of the calculator program 


@app.route('/math',methods=['POST'])
def math_ops():
    if(request.method == 'POST'):
        ops = request.form['operation']
        num1 = int(request.form['num1'])
        num2 = int(request.form['num2'])
        if ops == 'add':
            r = num1+num2
            result = "The sum of " + str(num1) + ' and ' + str(num2) + " is " + str(r)
        if ops == 'subtract':
            r = num1-num2
            result = "The subtract of " + str(num1) + ' and ' + str(num2) + " is " + str(r)
        if ops == 'multiply':
            r = num1*num2
            result = "The multiply of " + str(num1) + ' and ' + str(num2) + " is " + str(r)
        if ops == 'divide':
            r = num1/num2
            result = "The divide of " + str(num1) + ' and ' + str(num2) + " is " + str(r)
            
        return render_template('result.html' , result = result)
        
        ## The render_template here is used to access the the result.html page,
        ## the variable result is used to hold the value of the above user selected operation - add/subtract/multiply or divide.
        ## and results.html will be the user required and designed HTML page. 
        
        ## so the render_template will open the result.html with the output 
    
```
