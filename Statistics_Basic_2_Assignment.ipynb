#### Q1. What are the three measures of central tendency?

The central tendency of the dataset can be found out using the three important measures namely 

    1. mean
    2. median
    3. mode

2. The Measure of central tendancy is most important in EDA & Feature selection.
    
**MEAN**

1. The mean represents the average value of the dataset. 
2. It can be calculated as the sum of all the values in the dataset divided by the number of values


**MEDIAN**

1. Median is the middle value of the dataset in which the dataset is arranged in the ascending order or in descending order. 
2. When the dataset contains an even number of values, then the median value of the dataset can be found by taking the mean of the middle two values.


**MODE**

1. The mode represents the frequently occurring value in the dataset. 
2. Sometimes the dataset may contain multiple modes and in some cases, it does not contain any mode at all.

#### Q2. What is the difference between the mean, median, and mode? How are they used to measure the central tendency of a dataset?

A measure of central tendency is a single value that attempts to describe a set of data by identifying the central position within that set of data.

As such, measures of central tendency are sometimes called measures of central location.


**Mode**: It gives us the most frequently occuring value in a data set. 

**Median**: It is the middle number in an ordered dataset. If the dataset has odd number of values, then the median is the central value. If the dataset has even number of elements, then the median is given by calculating the average of the central 2 elements in the data set.

**Mean**: Is the sum of all values divided by the total number of values.

#### Q8. What do you understand about skewness in data?

The skewness in statistics is a measure of asymmetry or the deviation of a given random variables distribution from a symmetric distribution (like normal Distribution).

In Normal Distribution, we know that: Median = Mode = Mean

Skewness in statistics can be divided into two categories. They are:

1. Positive Skewness
2. Negative Skewness

**Positive Skewness**

The extreme data values are higher in a positive skew distribution, which increases the mean value of the data set. To put it another way, a positive skew distribution has the tail on the right side.

It means that, Mean > Median > Mode in positive skewness

**Negative Skewness**

The extreme data values are smaller in negative skewness, which lowers the datasets mean value. A negative skew distribution is one with the tail on the left side.

Hence, in negative Skewness, Mean <  Median < Mode.

#### Q10. Explain the difference between covariance and correlation. How are these measures used in statistical analysis?

**COVARIANCE**


1. Covariance is a statistical term that refers to a systematic relationship between two random variables in which a change in the other reflects a change in one variable.

2. The covariance value can range from - to +, with a negative value indicating a negative relationship and a positive value indicating a positive relationship.

3. The greater this number, the more reliant the relationship. Positive covariance denotes a direct relationship and is represented by a positive number.

4. A negative number, on the other hand, denotes negative covariance, which indicates an inverse relationship between the two variables. 

5. Covariance is great for defining the type of relationship, but it's terrible for interpreting the magnitude.

**Types of Covariance**

Covariance can be either positive or negative. It is classified into two sorts based on this:

**Positive Covariance**
1. A positive covariance between two variables indicates that they are heading in the same direction. 
2. The variables, in this case, behave similarly. That is, if the values of one variable (more or smaller) correspond to the values of another variable, they are said to be in positive covariance.

**Negative Covariance**
1. When two variables have a negative covariance, the variables shift in the opposite direction.
2. It is the inverse of positive covariance, in which higher values of one variable correlate to lower values of another and vice versa.

#### Q14. How do outliers affect measures of central tendency and dispersion? Provide an example.

**Outliers** are data points that are far from other data points. In other words, they're unusual values in a dataset. 

The most affected measure of central tendency is the MEAN.

```python
# Let us consider the below example of hieght of 25 people in class 10.
# Hieght in cms.
# This has no outlier.

height_of_class_10 = [140,144,141,124,133,132,133,149,176,165,177,180,165,134,165,170,171,172,176,175,166,167,155,159,175]
```

```python
# Now lets calculate the mean, median, mode, variance & std deviation.
```

```python
print("The mean height of class 10 is : ",np.mean( height_of_class_10))
print("The median  height of class 10 is : ",np.median( height_of_class_10))
print("The mode of  height of class 10 is : ",stats.mode( height_of_class_10)[0][0])
print("The variance of  height of class 10 is : ",np.var( height_of_class_10))
print("The standard deviation of height  class 10 is :",np.std( height_of_class_10))
```

```python
# Let us consider the below example of hieght of 25 people in class 10.
# Hieght in cms.
# with one outlier value as 590

height_of_class_10 = [140,144,141,124,133,132,133,149,176,165,177,180,165,134,165,170,171,172,176,175,166,167,155,159,590]
```

```python
print("The mean height of class 10 is : ",np.mean( height_of_class_10))
print("The median  height of class 10 is : ",np.median( height_of_class_10))
print("The mode of  height of class 10 is : ",stats.mode( height_of_class_10)[0][0])
print("The variance of  height of class 10 is : ",np.var( height_of_class_10))
print("The standard deviation of height  class 10 is :",np.std( height_of_class_10))
```

From the above 2 example what we can conclude is:
1. With an outlier (590 cms), the median and mode were unchanged, but we can see a huge difference in the MEAN value.
2. So we can conclude that as apart of the measure fo central tendancy the MEAN is highly affected with respect to the OUTLIERS.

3. As part of the measure of dispersion, both the variance & standard deviation has had huge impact as well due to the OUTLIER.
