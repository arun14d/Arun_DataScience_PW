#### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

1. Matplotlib is a low level graph plotting library in python that serves as a visualization utility.

2. Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias:
        
        import matplotlib.pyplot as plt
        
    
3. The plot() function is used to draw points (markers) in a diagram.
    1. By default, the plot() function draws a line from point to point.
    2. The function takes parameters for specifying points in the diagram.
        1. Parameter 1 is an array containing the points on the x-axis.
        2. Parameter 2 is an array containing the points on the y-axis.
    3. We can use the keyword argument marker to emphasize each point with a specified marker.
    4. We can use the keyword argument linestyle, or shorter ls, to change the style of the plotted line
4. We can use the xlabel() and ylabel() functions to set a label for the x- and y-axis.
5. We can use the grid() function to add grid lines to the plot.
6. With the subplot() function you can draw multiple plots in one figure:
    
#### Here are five plots that can be plotted using the Pyplot module of Matplotlib:

1. Line plot: A line plot is a simple plot that shows the relationship between two variables. It is the most basic type of plot, and it is often used to show trends over time.

2. Bar plot: A bar plot is a type of plot that shows the distribution of data. It is often used to compare different categories of data.

3. Scatter plot: A scatter plot is a type of plot that shows the relationship between two variables. It is often used to show the correlation between two variables.

4. Pie chart: A pie chart is a type of plot that shows the distribution of data. It is often used to show the percentage of a whole that is made up of different categories.

5. Histogram: A histogram is a type of plot that shows the distribution of data. It is often used to show the frequency of different values in a dataset.

```python
import matplotlib.pyplot as plt  # importing the pyplot module
```

#### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

    import numpy as np
    np.random.seed(3)
    x = 3 + np.random.normal(0, 2, 50)
    y = 3 + np.random.normal(0, 2, len(x))
    
    Note: Also add title, xlabel, and ylabel to the plot.

*ANSWER* : The Matplotlib module has a method for drawing scatter plots, it needs two arrays of the same length, one for the values of the x-axis, and one for the values of the y-axis.

#### Scatter plot: 

1. A scatter plot is a type of plot that shows the relationship between two variables. 
2. A scatter plot is a visual representation of how two variables relate to each other. 
3. We can use scatter plots to explore the relationship between two variables, for example by looking for any correlation between them.



```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
colours = np.random.rand(50)  
sizes = np.random.rand(50) * 700
plt.scatter(x,y,c=colours, s=sizes , alpha=1 ,marker='*')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot Example Chart ")
plt.show()
```

    import numpy as np

#### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

    import numpy as np
    company = np.array(["Apple", "Microsoft", "Google", "AMD"])
    profit = np.array([3000, 8000, 1000, 10000])

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

1. A bar plot is a type of plot that shows the distribution of data. 
2. It is often used to compare different categories of data.
3. Its typically used when we compare numerical data with the categorical data

```python
colour = ['Red','Green','Yellow','Blue']
plt.bar(company,profit,color=colour,width=.45 ,align='edge',edgecolor='black')   # for categorical V/s Numerical Data 
plt.xlabel("Name of the company")
plt.ylabel("Profits in Million $")
plt.title("Q3 Profit Results")
plt.grid()
plt.show()
```

```python
## The Horizontal bar plot representation

colour = ['Red','Green','Yellow','Blue']
plt.barh(company,profit,color=colour)    # for categorical V/s Numerical Data 
plt.ylabel("Name of the company")
plt.xlabel("Profits in Million $")
plt.title("Q3 Profit Results")
plt.grid()
plt.show()
```

#### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
    
    box1 = np.random.normal(100, 10, 200)
    box2 = np.random.normal(90, 20, 200)

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

1. The "Whisker Plot" is another name for the "Box Plot."
2. The whiskers are the two lines that run from the minimum to the lower quartile (the box's beginning) and then from the higher quartile (the box's end) to the maximum.
3. In the box plot, the box is drawn from the first to third quartiles, with a vertical line running through it at the median.
4. The x-axis of a box plot represents the data to be plotted, while the y-axis represents the frequency distribution.
5. Box plots are useful for visualizing the distribution of numerical values in a field. They come in handy for making comparisons across categorical variables and spotting outliers, if any exist in a dataset.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.figure(figsize=(12,6))

plt.subplot(1, 2, 1)
plt.boxplot(box1)
plt.title("BoxPlot Chart - 1")

plt.subplot(1, 2, 2)
plt.boxplot(box2)
plt.title("BoxPlot Chart - 2")

plt.show()
```
