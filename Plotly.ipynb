```python
import seaborn as sns
import plotly.graph_objects as go
```

```python
## Q1. Load the "titanic" dataset using the load_dataset function of seaborn.
##    Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.
```

```python
# loading the titanic dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=titanic['age'],y = titanic['fare'], mode='markers'))
fig.show()
```

```python
# Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
fig = go.Figure()
fig.add_trace(go.Box(x=tips['sex'],y = tips['total_bill']))
fig.show()

## box plot of tips['sex'] v/s tips['total_bill']
## A Box Plot is also known as Whisker plot is created to display the summary of the set of data values
## having properties like minimum, first quartile, median, third quartile and maximum.
```

```python
# Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill"
## column in the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
import plotly.express as px

tips = sns.load_dataset('tips')
fig = px.histogram(tips, x="sex", y="total_bill", color="sex", pattern_shape="smoker")
fig.show()
```

```python
# Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for the color parameter.
# Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the dimensions parameter.
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
fig = px.scatter_matrix(iris,
    dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],
    color="species")
fig.show()
```
