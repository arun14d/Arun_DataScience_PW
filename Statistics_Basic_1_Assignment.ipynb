#### Q2. Define the different types of statistics and give an example of when each type might be used.

Basically, there are two types of statistics.

1. Descriptive Statistics
2. Inferential Statistics


*Descriptive Statistics*
---

1. The data is summarised and explained in descriptive statistics. 
2. The summarization is done from a population sample utilising several factors such as mean and standard deviation. 
3. Descriptive statistics is a way of organising, representing, and explaining a set of data using charts, graphs, and summary measures. 
4. Histograms, pie charts, bars, and scatter plots are common ways to summarise data and present it in tables or graphs. 
5. Descriptive statistics are just that: descriptive. They don’t need to be normalised beyond the data they collect.


Eg: - Calculating the average height of the class 10 students from school - Acts Secondary School. --> we use mean

*Inferential Statistics*
---

1. We attempt to interpret the meaning of descriptive statistics using inferential statistics.
2. We utilise inferential statistics to convey the meaning of the collected data after it has been collected, evaluated, and summarised. 
3. The probability principle is used in inferential statistics to determine if patterns found in a study sample may be extrapolated to the wider population from which the sample was drawn. 
4. Inferential statistics are used to test hypotheses and study correlations between variables, and they can also be used to predict population sizes.
5. Inferential statistics are used to derive conclusions and inferences from samples, i.e. to create accurate generalisations.

Eg:- Exit poll of the state election

#### Q5. Explain the concept of levels of measurement and give an example of a variable for each level.

There are 4 level of measurement of data.
1. Nominal Scale Data
2. Ordinal Scale data
3. Interval Scale Data
4. Ratio Scale Data

1. Nominal Scale Data

    * Qualitative or Categorical variable are involved.
    * Eg:- Gender, Colors, Labels -> Fixed set of categories
    * Order does not matter.
    * We focus more on count and more on distribution of data
    
      Survery of 10 people --> to find best color amony the 3 colors - Red/Blue/Yellow
      
        Red - 5
        Blue - 3
        Yellow - 2

        Clearly the Red has 5 count, so we can say 50% of the people think that RED is the best color.

2. Ordinal Scale Data

    * Ranking and order does matter
    * Difference cannot be measured
    
    Ex: Qualifications
        PhD - 1st Rank
        Masters -  2nd Rank
        BE - 3rd Rank
        BSc - 4th Rank
        Bcom - 5th Rank
    
    * This are order, and difference cannot be calculated
    
    Ex : Marks: 100 - 1st Rank, 90 - 2nd Rank, 65 - 4th Rank, 70 - 3rd Rank, 30 - 5th Rank
    
    In case of rank, its ordinal, but in case of Marks, its not a ordinal scale data
    
    
3. Interval Scale Data

    * The Rank & order matters.
    * Difference can be measured
    * Ratio cannot be measured.
    * Does not have 0 starting value. --> only 0 is not the start, it can have -ve temperature as well.
    
    Ex:- Temperature
    30 F
    60 F
    80 F
    90 F
    
    Difference can be measured for the above example, it can be ordered and ranked as well.
    
4. Ratio Scale Data

    * Order & Rank matters.
    * Differences & Ratio are measurable.
    * It does have a 0 Starting point.
    
    Ex: Grades - it has 0 as starting point. It can't have negative.
    
    100,90,60,35,45,50

#### Q6. Why is it important to understand the level of measurement when analyzing data? Provide an example to illustrate your answer.

1. The level at which you measure a variable determines how you can analyze your data.

2. The different levels limit which descriptive statistics you can use to get an overall summary of your data, and which type of inferential statistics you can perform on your data to support or refute your hypothesis.

3. In many cases, your variables can be measured at different levels, so you have to choose the level of measurement you will use before data collection begins.

**Example of a variable at 2 levels of measurement**

1. You can measure the variable of income at an ordinal or ratio level.
2. Ordinal level: You create brackets of income ranges: $0–$19,999, $20,000–$39,999, and $40,000–$59,999. You ask participants to select the bracket that represents their annual income. The brackets are coded with numbers from 1–3.
3. Ratio level: You collect data on the exact annual incomes of your participants.

|Participant|Income (ordinal level)|Income (ratio level)|
---|---|---|
A|Bracket 1|$12550
B|Bracket 2|$39700 
C|Bracket 3|$40300

1. At a ratio level, you can see that the difference between A and B’s incomes is far greater than the difference between B and C’s incomes.

2. At an ordinal level, however, you only know the income bracket for each participant, not their exact income. Since you cannot say exactly how much each income differs from the others in your data set, you can only order the income levels and group the participants.

#### Q10. What are some common measures of central tendency and variability used in statistics? Explain how each measure can be used to describe a dataset.

1. The central tendency of the dataset can be found out using the three important measures namely 
    1. mean
    2. median
    3. mode

2. The Measure of central tendancy is most important in EDA & Feature selection.
    
**MEAN**

1. The mean represents the average value of the dataset. 
2. It can be calculated as the sum of all the values in the dataset divided by the number of values


**MEDIAN**

1. Median is the middle value of the dataset in which the dataset is arranged in the ascending order or in descending order. 
2. When the dataset contains an even number of values, then the median value of the dataset can be found by taking the mean of the middle two values.


Consider the given dataset with the odd number of observations arranged in descending order – 23, 21, 18, 16, 15, 13, 12, 10, 9, 7, 6, 5, and 2

Here 12 is the middle or median number that has 6 values above it and 6 values below it.


Now, consider another example with an even number of observations that are arranged in descending order – 40, 38, 35, 33, 32, 30, 29, 27, 26, 24, 23, 22, 19, and 17


When you look at the given dataset, the two middle values obtained are 27 and 29.

Now, find out the mean value for these two numbers.

i.e.,(27+29)/2 =28

Therefore, the median for the given data distribution is 28.


**MODE**

1. The mode represents the frequently occurring value in the dataset. 
2. Sometimes the dataset may contain multiple modes and in some cases, it does not contain any mode at all.

Consider the given dataset 5, 4, 2, 3, 2, 1, 5, 4, 5


Since the mode represents the most common value. Hence, the most frequently repeated value in the given dataset is 5.


Based on the properties of the data, the measures of central tendency are selected.

1. If you have a symmetrical distribution of continuous data, all the three measures of central tendency hold good. But most of the times, the analyst uses the mean because it involves all the values in the distribution or dataset.
2. If you have skewed distribution, the best measure of finding the central tendency is the median.
3. If you have the original data, then both the median and mode are the best choice of measuring the central tendency.
4. If you have categorical data, the mode is the best choice to find the central tendency.
