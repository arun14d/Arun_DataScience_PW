#### Q1. What is a database? Differentiate between SQL and NoSQL databases.

1. A database is an organized collection of structured information, or data, typically stored electronically in a computer system. 

2. A database is usually controlled by a database management system (DBMS). Together, the data and the DBMS, along with the applications that are associated with them, are referred to as a database system, often shortened to just database.

3. Data within the most common types of databases in operation today is typically modeled in rows and columns in a series of tables to make processing and data querying efficient. 

4. The data can then be easily accessed, managed, modified, updated, controlled, and organized. Most databases use structured query language (SQL) for writing and querying data.

---------------------

| Difference B/W | SQL | NOSQL |
| --- | --- | --- |
| 1. | Databases are categorized as Relational Database Management System (RDBMS). | NoSQL databases are categorized as Non-relational or distributed database system. |
| 2. | SQL databases have fixed or static or predefined schema. | NoSQL databases have dynamic schema.|
| 3. | SQL databases display data in form of tables so it is known as table-based database. | NoSQL databases display data as collection of key-value pair, documents, graph databases or wide-column stores.|
| 4. | SQL databases are vertically scalable. | NoSQL databases are horizontally scalable.|
| 5. | SQL databases use a powerful language "Structured Query Language" to define and manipulate the data. | In NoSQL databases, collection of documents are used to query the data. It is also called unstructured query language. It varies from database to database. |
| 6. | SQL databases are best suited for complex queries | NoSQL databases are not so good for complex queries because these are not as powerful as SQL queries.|
| 7. | SQL databases are not best suited for hierarchical data storage. | NoSQL databases are best suited for hierarchical data storage. |
| 8. | MySQL, Oracle, Sqlite, PostgreSQL and MS-SQL etc. are the example of SQL database. | MongoDB, BigTable, Redis, RavenDB, Cassandra, Hbase, Neo4j, CouchDB etc. are the example of nosql database |

#### Q2. What is DDL? Explain why CREATE, DROP, ALTER, and TRUNCATE are used with an example.

1. Data Definition Language(DDL) is used to define database schema in DBMS.

2. Data Definition Language(DDL) commands in DBMS are used to define the database objects. We can create, delete and alter tables using DDL commands.

3. Since a database structure is designed by DDL commands, they are called Data Definition Languages (DDLs). They consist of commands such as CREATE, ALTER, TRUNCATE, and DROP.

4. DDL in DBMS is a language that allows user to define the database components and their relationship with each other.

5. These commands work with the structure of tables like creating a table, deleting a table, and altering a table.

6. DDL commands allow us to define and change the data types of table columns, and integrity constraints while defining the table structure. Integrity constraints are rules set for the data being stored in the table. This is explained in the next section of this article.

7. All DDL commands are auto-committed which means the changes done using them are permanently saved in the Database but they can be rolled back.

#### Q3. What is DML? Explain INSERT, UPDATE, and DELETE with an example.

#### **DML**
1. DML is a language that is used to manipulate data stored in the database.

2. These commands are used to manipulate table structures in a database by updating or deleting, or retrieving existing data or inserting new data into them.

3. DML commands are not auto-committed, which means the changes done using these commands are not saved into the database automatically.

4. Users need to commit them manually. Changes done using DML commands in SQL can be rolled back.

##### **INSERT**

    1. Insert command is used to insert data into a table.
    2. When a table is created, we need to add data to it, this data is added using the INSERT DML command

    
    Syntax:- 
    
    INSERT INTO table_name(Columns_list) values (column values);


```python
## Sample python program to insert 2 rows into the table - test1.test_table.
## The 2 insert statement will insert 2 rows into the the table.

import mysql.connector
mydb = mysql.connector.connect(
  host="localhost",
  user="abc",
  password="password"
)
mycursor = mydb.cursor()
mycursor.execute("insert into test1.test_table values(123,'Arun',344,99.9,'RCB')") ## Insert statement 1
mycursor.execute("insert into test1.test_table values(124,'virat',345,99,'RCB')") ## ## Insert statement 1
mydb.commit() ## always call when insert statement, to commit after the insert is done.
mydb.close()
```

#### **UPDATE**

    1. The UPDATE statement is used to modify the existing records in a table.
    2. The WHERE clause specifies which record(s) that should be updated. If we omit the WHERE clause, all records in the table will be updated!
    
    **Syntax**
    
    UPDATE table_name
    SET column1 = value1, column2 = value2, ...
    WHERE condition;
    

```python
## Example for updating single column
import mysql.connector
mydb = mysql.connector.connect(
  host="localhost",
  user="abc",
  password="password"
)
mycursor = mydb.cursor()
mycursor.execute("update Arun_sample_DB.Student_details set Name = 'Arun Kumar' where stud_id = 123") 
mydb.commit() ## always call when update statement, to commit after the update is done.
mydb.close()
```

##### **DELETE**

    1.The DELETE statement is used to delete existing records in a table.
    2. The WHERE clause specifies which record(s) should be deleted. If we omit the WHERE clause, all records in the table will be deleted!
    
    Syntax
    
    DELETE FROM table_name WHERE condition;
    
    

```python
## Example for deleting a record

import mysql.connector
mydb = mysql.connector.connect(
  host="localhost",
  user="abc",
  password="password"
)
mycursor = mydb.cursor()
## The below sql statement deletes the record with stud_id - 123
mycursor.execute("delete from  Arun_sample_DB.Student_details where stud_id = 123") 
mydb.commit() ## always call when update statement, to commit after the update is done.
mydb.close()
```

##### Q5. Explain Primary Key and Foreign Key.

##### **PRIMARY KEY**
    1. A primary key is a column (or set of columns) in a table that uniquely identifies each row in the table. 
    2. It cannot contain null values and must be unique across all rows in the table. Only one primary key is allowed in a table.
    3. A primary key is basically a combination of the 'UNIQUE' and 'Not Null' constraints. 
    4. Thus, it cannot be a NULL value. Another important point to be noted about primary key is that its value cannot be deleted from the parent table.

    CREATE TABLE Persons (
        ID int NOT NULL, --> Primary key that will be defined  in a table, shouldnt be NULL.
        LastName varchar(255) NOT NULL,
        FirstName varchar(255),
        Age int,
        PRIMARY KEY (ID)   --> We by doin this, making the ID column a primary key of the table PERSONS.
    );

##### **FOREIGN KEY**
    1. A foreign key is a column (or set of columns) in a table that refers to the primary key in another table. 
    2. It is used to establish a link between the two tables and is used to enforce referential integrity in the database. Foreign key is basically the    field/column in a table that is analogous to the primary key of other table.
    3. Unlike a primary key, a table can have more than one foreign key. Also, the foreign key can contain duplicate and null values in a relational database. The value of a foreign key can be deleted from the child table.

    CREATE TABLE Orders (
        OrderID int NOT NULL,
        OrderNumber int NOT NULL,
        PersonID int,
        PRIMARY KEY (OrderID), --> Primary key of the table ORders.
        FOREIGN KEY (PersonID) REFERENCES Persons(ID) --> PersonID frrom Order table is refering the ID column from Person table.
);
