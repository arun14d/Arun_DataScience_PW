#### Q3. What is App routing in Flask? Why do we use app routes?

1. App routing is the technique used to map the specific URL with the associated function intended to perform some task. 
2. The Latest Web frameworks use the routing technique to help users remember application URLs.

3. It is helpful to access the desired page directly without navigating from the home page.
4. The URL (‘/’) is linked with the root URL in every application.  

```python
## sample example of route()
from flask import Flask
from flask import request
## create an object of Class Flask..
app = Flask(__name__)
## decorate the custom function
## if we are calling via Function name - 
## route is used to expose the below fucntion on url
## / -> route for a home location
@app.route("/HelloWorld")
def hello_world():
    return "Hello, World!"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

#### Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

##### **url_for()** 

    1. To build a URL to a specific function, use the url_for() function.
    2. It accepts the name of the function as its first argument and any number of keyword arguments, each corresponding to a variable part of the URL rule.

```python
## importing required modules
from flask import Flask, redirect, url_for
app = Flask(__name__)

## Admin func 
@app.route('/admin')  
def admin():  
    return 'You have admin access'  
  
@app.route('/librarian')  
def librarion():  
    return 'You have librarian access'  
  
@app.route('/student')  
def student():  
    return 'You have student access'  
  
@app.route('/user/<name>')  
def user(name):  
    if name == 'admin':  
        return redirect(url_for('admin'))  
    if name == 'librarian':  
        return redirect(url_for('librarian'))  
    if name == 'student':  
        return redirect(url_for('student'))  
if __name__=="__main__":
    app.run(host="0.0.0.0")
    
```

For example, 
1. the URL https://brown-lawyer-fwegw.pwskills.app:5000/user/admin is redirected to the URL https://brown-lawyer-fwegw.pwskills.app:5000/admin, 
2. the URL https://brown-lawyer-fwegw.pwskills.app:5000/user/librarion, is redirected to the URL https://brown-lawyer-fwegw.pwskills.app:5000/librarion, 
3. the URL https://brown-lawyer-fwegw.pwskills.app:5000/user/student is redirected to the URL https://brown-lawyer-fwegw.pwskills.app:5000/student.

1. simulates the library management system which can be used by the three types of users, i.e., admin, librarion, and student.
2. There is a specific function named user() which recognizes the user the redirect the user to the exact function which contains the implementation for this particular function.

**Benefits of the Dynamic URL Building**
1. It avoids hard coding of the URLs.
2. We can change the URLs dynamically instead of remembering the manually changed hard-coded URLs.
3. URL building handles the escaping of special characters and Unicode data transparently.
4. The generated paths are always absolute, avoiding unexpected behavior of relative paths in browsers.
5. If your application is placed outside the URL root, for example, in /myapplication instead of /, url_for() properly handles that for you.
